```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import plotnine as p9
import seaborn as sns
%pylab inline
```

```python
# read in df_paiso, containing all 272526 CCS reads from all paiso seq results.
df_paiso = pd.read_csv("../raw/PAIsoSeq_Wu_2022.csv",index_col=0)
df_paiso

df_paiso['sampleID'] = df_paiso['Genotype']+'_' + df_paiso['oocyte_ID'].astype(str)
df_paiso
# only need the ctr5 6 7 8 and cKO 5 6 7 8 oocytes
df_paiso = df_paiso[df_paiso['oocyte_ID'].isin([5,6,7,8])]
df_paiso
```

```python
ntop = 100
# to study the PC1 ntop genes
# get the PC1 genes:
df_PC = pd.read_csv("../../results/PC1_from_rM_seq.csv")
df_PC
list_top = df_PC.loc[:ntop,'Unnamed: 0'].tolist()
list_top

# visualize the ntop genes:
df_paiso_ntop = df_paiso[df_paiso['gene_ID'].isin(list_top)]
df_paiso_ntop
```

```python
# group
info_df_paiso_ntop = df_paiso_ntop.groupby(['gene_ID','sampleID']).count().reset_index()
info_df_paiso_ntop = info_df_paiso_ntop[['gene_ID','sampleID','CCS_ID']]

info_df_paiso_ntop.columns = ['gene_ID','sampleID','count']
# info_df_paiso_ntop
```

```python
# use the overall count to normalize
# the count of transcripts in each sample:
countAll = df_paiso.groupby('sampleID').count().reset_index()
countAll = countAll[['sampleID','CCS_ID']]
countAll.columns = ['sampleID','countAll']
# countAll
```

```python
# merge info_df_paiso_ntop and countAll
info_df_paiso_ntop_merge = pd.merge(left=info_df_paiso_ntop, right = countAll,
                                   left_on='sampleID',right_on='sampleID')
info_df_paiso_ntop_merge['norm_count']=info_df_paiso_ntop_merge['count']/info_df_paiso_ntop_merge['countAll']
# sort
info_df_paiso_ntop_merge = info_df_paiso_ntop_merge.sort_values(['sampleID','count'])
# info_df_paiso_ntop_merge
```

```python
# plot the norm count
title = 'info_df_paiso_ntop_merge'
plot = (
    p9.ggplot(info_df_paiso_ntop_merge, 
                  p9.aes(x='gene_ID',y='norm_count'))
#     + p9.geom_boxplot(outlier_shape='',alpha=0.3)
    + p9.geom_point(alpha=0.3)
#     + p9.geom_point()
    + p9.theme(void)
    + p9.labs(title = title)
    + p9.facet_grid('sampleID~')
    + p9.theme(axis_text_x=p9.element_text(angle=90))
)
plot
```

```python
# info_df_paiso_ntop_merge to wide:
info_df_paiso_ntop_merge_wide = info_df_paiso_ntop_merge.pivot(index='gene_ID',columns='sampleID',values='norm_count')
info_df_paiso_ntop_merge_wide=info_df_paiso_ntop_merge_wide.reset_index()
info_df_paiso_ntop_merge_wide=info_df_paiso_ntop_merge_wide.fillna(0)
# sort
info_df_paiso_ntop_merge_wide=info_df_paiso_ntop_merge_wide.sort_values(['cKO_5','cKO_6','cKO_7','cKO_8'],ascending=False)

sns.heatmap(info_df_paiso_ntop_merge_wide.iloc[:,1:],cmap="Blues")
plt.savefig("../results/heterogeneity/topGenes"+str(ntop)+".pdf")
```
