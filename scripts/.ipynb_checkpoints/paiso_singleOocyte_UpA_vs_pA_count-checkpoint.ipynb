```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import plotnine as p9
%pylab inline
```

```python
# to study the the U-polyA vs polyA ratio in each oocyte.
```

```python
# read in df_paiso, containing all 272526 CCS reads from all paiso seq results.
df_paiso = pd.read_csv("../raw/PAIsoSeq_Wu_2022.csv",index_col=0)
df_paiso
```

```python
# transcript number in each oocyte:
info_paiso = df_paiso.groupby(['Genotype','oocyte_ID','type']).count().reset_index()
info_paiso = info_paiso[['Genotype','oocyte_ID','type','CCS_ID']]
info_paiso.columns = ['Genotype','oocyte_ID','type','count']
info_paiso_U = info_paiso[info_paiso['type']=='U_polyA_RNA']
info_paiso_pa = info_paiso[info_paiso['type']=='polyA_RNA']

info_paiso_U
info_paiso_pa

info_paiso2 = pd.merge(left=info_paiso_U, right=info_paiso_pa, 
                       left_on=['Genotype','oocyte_ID'],
                      right_on=['Genotype','oocyte_ID'])
info_paiso2['Uvspa']= info_paiso2['count_x']/info_paiso2['count_y']
info_paiso2['id'] = info_paiso2['Genotype'] + '_' + info_paiso2['oocyte_ID'].astype('str')
info_paiso2['sum'] = info_paiso2['count_x'] + info_paiso2['count_y']
info_paiso2['UpApercent'] = info_paiso2['count_x']/info_paiso2['sum']

thre=1000
info_paiso2 = info_paiso2[info_paiso2['sum']>thre]
info_paiso2
```

```python
# plot the ratio
title = 'UpApercent'
plot = (
    p9.ggplot(info_paiso2, 
                  p9.aes(x='Genotype',y='UpApercent'))
    + p9.geom_boxplot(outlier_shape='')
    + p9.geom_jitter(p9.aes(color='sum'))
    + p9.theme(void)
    + p9.labs(title = title)
#     + p9.xlim(0,250) # ylim 200, more than 95% data are included
)
plot
plot.save("../results/heterogeneity/" + title + ".pdf", height=2, width=2)
```
