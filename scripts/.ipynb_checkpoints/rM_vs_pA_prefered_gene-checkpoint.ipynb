```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import plotnine as p9
%pylab inline
```

```python
# To get rM_prefered and pA_prefered genes from the rM vs pA.
# Output: plot the polyA len of the two groups of genes
```

```python
# this is to plot the polyA len ranges of all genes from paiso-seq
# read in paiso df, 272526. neglect the up/dn
df_paiso = pd.read_csv("../raw/PAIsoSeq_Wu_2022.csv",index_col=0)
df_paiso
```

```python
# read in teh rM pA corr df:
df_rM_pA_GV_short = pd.read_csv("../raw/RNAseq/df_rM_pA_GV_short.csv",index_col=0)
df_rM_pA_GV_short
```

```python
# use df_rM_pA_GV_short to get rM_prefered and pA_prefered genes:
df_rM_pA_GV_short['rM_minus_pA'].describe()

# use y=x+2 and y=x-2 to define the difference.

# so rM_prefered are 2, pA prefered are -2
th_rM = 2
th_pA = -2

rM_prefered = df_rM_pA_GV_short[df_rM_pA_GV_short['rM_minus_pA']>th_rM]
pA_prefered = df_rM_pA_GV_short[df_rM_pA_GV_short['rM_minus_pA']<th_pA]

pA_prefered #211
rM_prefered #3128
```

```python
rM_prefered_cKO = df_paiso[(df_paiso['Genotype']=='cKO')&(df_paiso['gene_ID'].isin(rM_prefered['gene_id']))]
rM_prefered_cKO['prefered'] = 'rM_prefered_cKO'

rM_prefered_ctr = df_paiso[(df_paiso['Genotype']=='ctr')&(df_paiso['gene_ID'].isin(rM_prefered['gene_id']))]
rM_prefered_ctr['prefered'] = 'rM_prefered_ctr'


pA_prefered_cKO = df_paiso[(df_paiso['Genotype']=='cKO')&(df_paiso['gene_ID'].isin(pA_prefered['gene_id']))]
pA_prefered_cKO['prefered'] = 'pA_prefered_cKO'

pA_prefered_ctr = df_paiso[(df_paiso['Genotype']=='ctr')&(df_paiso['gene_ID'].isin(pA_prefered['gene_id']))]
pA_prefered_ctr['prefered'] = 'pA_prefered_ctr'


pd_merge = pd.concat([rM_prefered_cKO,rM_prefered_ctr,pA_prefered_cKO,pA_prefered_ctr])
pd_merge
```

```python
pd_merge.to_csv("../results/rM_pA_prefered_gene_list.csv")
```

```python
# the percentage of U_polyA and polyA RNA:
info_pd_merge = pd_merge.groupby(['prefered','type']).count().reset_index()
info_pd_merge
info_pd_merge.to_csv("../results/info_rM_vs_pA_prefered_gene.csv")
```
