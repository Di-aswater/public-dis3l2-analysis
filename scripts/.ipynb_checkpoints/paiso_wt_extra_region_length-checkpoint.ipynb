```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import plotnine as p9
%pylab inline
```

```python
# read in df_paiso, containing all 272526 CCS reads from all paiso seq results.
df_paiso = pd.read_csv("../raw/PAIsoSeq_Wu_2022.csv",index_col=0)
df_paiso
```

```python
# only select ctr group:
df_paiso_wt = df_paiso[df_paiso['Genotype']=='ctr']
df_paiso_wt

# get percentage of polyA_RNA and U_polyA_RNA
df_paiso_wt.groupby('type').count()

# U_polyA_RNA: 4665/(4665+147058) = 3%
# polyA_RNA: 147058/(4665+147058) = 97%
```

```python
# percentage of the original types of transcripts: 
df_paiso_wt.groupby('compare_pA_tail').count()
# 15% equal, 3% longer_tail, 82% longer_pA
```

```python
# calculate the length of extra region:
df_paiso_wt['extra'] = df_paiso_wt['polyA_len'] - df_paiso_wt['tail_len']
df_paiso_wt['type'] = df_paiso_wt['type'].astype('category')
df_paiso_wt['type'] = df_paiso_wt['type'].cat.reorder_categories(['U_polyA_RNA','polyA_RNA'])

df_paiso_wt['extra'] = abs(df_paiso_wt['extra'])

# the features of U_polyA_RNA:
df_paiso_wt.groupby('type')['extra'].describe(percentiles =[.95]) # 95%, length=30.

# get mean value
info_df_paiso_wt = df_paiso_wt.groupby('type')['extra'].describe().reset_index()
info_df_paiso_wt
```

```python
# plot the length of the internal extra region:
title = 'length_of_extra_region'
plot = (
    p9.ggplot(df_paiso_wt, 
                  p9.aes(x='type', y='extra'))
    + p9.geom_boxplot(outlier_color='gray',width=0.1)
    + p9.theme(void)
    + p9.labs(title = title)
    + p9.ylim(0,30) # ylim 30, more than 95% data are included
    + p9.geom_hline(yintercept = 0, linetype='dashed')
    + p9.geom_hline(yintercept = 1, linetype='dashed')
    + p9.geom_hline(yintercept = 3, linetype='dashed')
)
plot
plot.save("../results/paiso_wt_" + title + ".pdf", height=6, width=4)
```

```python
# plot the length distribution:
title = 'density_of_length'
plot = (
    p9.ggplot(df_paiso_wt, 
                  p9.aes(x='extra'))
    + p9.geom_density(p9.aes(color='type'))
    + p9.theme(void)
    + p9.labs(title = title)
    + p9.xlim(0,30) # ylim 30, more than 95% data are included
)
plot
# plot.save("../results/paiso_wt_" + title + ".pdf", height=6, width=4)
```
