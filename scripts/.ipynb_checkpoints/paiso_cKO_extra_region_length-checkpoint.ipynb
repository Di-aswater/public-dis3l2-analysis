```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import plotnine as p9
%pylab inline
```

```python
# read in df_paiso, containing all 272526 CCS reads from all paiso seq results.
df_paiso = pd.read_csv("../raw/PAIsoSeq_Wu_2022.csv",index_col=0)
df_paiso
```

```python
# only select one genotype:
df_paiso_cKO = df_paiso[df_paiso['Genotype']=='cKO']
df_paiso_cKO

# get percentage of polyA_RNA and U_polyA_RNA
df_paiso_cKO.groupby('type').count()

# U_polyA_RNA: 44471/(44471+76332) = 37%
# polyA_RNA: 76332/(44471+76332) = 63%
```

```python
# calculate the length of extra region:
df_paiso_cKO['extra'] = df_paiso_cKO['polyA_len'] - df_paiso_cKO['tail_len']
df_paiso_cKO['type'] = df_paiso_cKO['type'].astype('category')
df_paiso_cKO['type'] = df_paiso_cKO['type'].cat.reorder_categories(['U_polyA_RNA','polyA_RNA'])

df_paiso_cKO['extra'] = abs(df_paiso_cKO['extra'])

# the features of U_polyA_RNA:
df_paiso_cKO.groupby('type')['extra'].describe(percentiles =[.95]) # 95%, length=43.
```

```python
# plot the length of the internal extra region:
title = 'length_of_extra_region'
plot = (
    p9.ggplot(df_paiso_cKO, 
                  p9.aes(x='type', y='extra'))
    + p9.geom_boxplot(outlier_color='gray',width=0.1)
    + p9.theme(void)
    + p9.labs(title = title)
    + p9.ylim(0,100) # ylim 30, more than 95% data are included
    + p9.geom_hline(yintercept = 0, linetype='dashed')
    + p9.geom_hline(yintercept = 1, linetype='dashed')
    + p9.geom_hline(yintercept = 24, linetype='dashed')
)
plot
plot.save("../results/paiso_cKO_" + title + ".pdf", height=6, width=4)
```

```python
# plot the length distribution:
title = 'density_of_length'
plot = (
    p9.ggplot(df_paiso_cKO, 
                  p9.aes(x='extra'))
    + p9.geom_density(p9.aes(color='type'))
    + p9.theme(void)
    + p9.labs(title = title)
    + p9.xlim(0,30) # ylim 30, more than 95% data are included
)
plot
# plot.save("../results/paiso_wt_" + title + ".pdf", height=6, width=4)
```
