```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import plotnine as p9
%pylab inline
```

```python
# read in df_paiso, containing all 272526 CCS reads from all paiso seq results.
df_paiso = pd.read_csv("../raw/PAIsoSeq_Wu_2022.csv",index_col=0)
df_paiso
```

```python
# only select ctr group:
df_paiso_cKO = df_paiso[df_paiso['Genotype']=='cKO']
df_paiso_cKO
```

```python
polyA_RNA = df_paiso_cKO[df_paiso_cKO['type']=='polyA_RNA']
U_polyA_RNA = df_paiso_cKO[df_paiso_cKO['type']=='U_polyA_RNA']

# for polyA_RNA: the extra length.
polyA_RNA['extra_length'] = polyA_RNA['polyA_len'] -polyA_RNA['tail_len']
polyA_RNA['extra_length_start_site'] =0

# only extract the most 5'end from pA_seq to get the exceeded
polyA_RNA['extra_region_in_pA'] = [x[y:z]for x,y,z in zip(polyA_RNA.pA_seq, polyA_RNA.extra_length_start_site, polyA_RNA.extra_length)]
polyA_RNA

# for polyA_RNA: the extra length.
U_polyA_RNA['extra_length'] = U_polyA_RNA['tail_len'] -U_polyA_RNA['polyA_len']
U_polyA_RNA['extra_length_start_site'] =0

# only extract the most 5'end from pA_seq to get the exceeded
U_polyA_RNA['extra_region_in_UpA'] = [x[y:z]for x,y,z in zip(U_polyA_RNA.tail_seq, U_polyA_RNA.extra_length_start_site, U_polyA_RNA.extra_length)]
U_polyA_RNA
polyA_RNA
```

```python
# polyA_RNA extra region base composition:
polyA_RNA['U_percent_in_extra']= polyA_RNA.extra_region_in_pA.str.count('T')/polyA_RNA['extra_length']
polyA_RNA['A_percent_in_extra']= polyA_RNA.extra_region_in_pA.str.count('A')/polyA_RNA['extra_length']
polyA_RNA['C_percent_in_extra']= polyA_RNA.extra_region_in_pA.str.count('C')/polyA_RNA['extra_length']
polyA_RNA['G_percent_in_extra']= polyA_RNA.extra_region_in_pA.str.count('G')/polyA_RNA['extra_length']
# drop na values:
polyA_RNA=polyA_RNA.dropna()
```

```python
# U_polyA_RNA extra region base composition:
U_polyA_RNA['U_percent_in_extra']= U_polyA_RNA.extra_region_in_UpA.str.count('T')/U_polyA_RNA['extra_length']
U_polyA_RNA['A_percent_in_extra']= U_polyA_RNA.extra_region_in_UpA.str.count('A')/U_polyA_RNA['extra_length']
U_polyA_RNA['C_percent_in_extra']= U_polyA_RNA.extra_region_in_UpA.str.count('C')/U_polyA_RNA['extra_length']
U_polyA_RNA['G_percent_in_extra']= U_polyA_RNA.extra_region_in_UpA.str.count('G')/U_polyA_RNA['extra_length']

U_polyA_RNA
```

```python
title = 'base composition of extra region'
use = U_polyA_RNA
title2 = 'U_polyA_RNA'

plot = (
    p9.ggplot()
    + p9.geom_boxplot(use, 
                  p9.aes(x=0, y='U_percent_in_extra'),size=2)
    + p9.geom_boxplot(use, 
                  p9.aes(x=1, y='A_percent_in_extra'),size=2)
    + p9.geom_boxplot(use, 
                  p9.aes(x=2, y='C_percent_in_extra'),size=2)
    + p9.geom_boxplot(use, 
                  p9.aes(x=3, y='G_percent_in_extra'),size=2)
    + p9.geom_jitter(use, 
                  p9.aes(x=0, y='U_percent_in_extra'),color='green',alpha=0.01)
    + p9.geom_jitter(use, 
                  p9.aes(x=1, y='A_percent_in_extra'),color='green',alpha=0.01)
    + p9.geom_jitter(use, 
                  p9.aes(x=2, y='C_percent_in_extra'),color='green',alpha=0.01)
    + p9.geom_jitter(use, 
                  p9.aes(x=3, y='G_percent_in_extra'),color='green',alpha=0.01)

    + p9.theme(void)
    + p9.xlab('U, A, C, G')
    + p9.ylab('base percentage')
    + p9.labs(title = title)
)
plot
plot.save("../results/paiso_cKO_basePentageExtra_" + title2 + ".tiff", height=4, width=4)
```
