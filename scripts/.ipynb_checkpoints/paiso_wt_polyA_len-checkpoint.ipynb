```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import plotnine as p9
%pylab inline
```

```python
# read in df_paiso, containing all 272526 CCS reads from all paiso seq results.
df_paiso = pd.read_csv("../raw/PAIsoSeq_Wu_2022.csv",index_col=0)
df_paiso
```

```python
# only select ctr group:
df_paiso_wt = df_paiso[df_paiso['Genotype']=='ctr']
df_paiso_wt
```

```python
# plot the length of the internal extra region in U_polyA_RNA:
title = 'tail_len'
plot = (
    p9.ggplot(df_paiso_wt, 
                  p9.aes(x='tail_len',color='type'))
#     + p9.geom_boxplot(outlier_color='gray',width=0.1)
    + p9.geom_density()
    + p9.theme(void)
    + p9.labs(title = title)
    + p9.xlim(0,250) # ylim 200, more than 95% data are included
)
plot
plot.save("../results/paiso_wt_" + title + ".pdf", height=2, width=2)
```

```python
# plot the fq_len:
df_paiso_wt['fq_len'] = df_paiso_wt['fq_seq'].str.len()
title = 'fq_len'
plot = (
    p9.ggplot(df_paiso_wt, 
                  p9.aes(x='fq_len',color='type'))
#     + p9.geom_boxplot(outlier_color='gray',width=0.1)
    + p9.geom_density()
    + p9.theme(void)
    + p9.labs(title = title)
#     + p9.xlim(0,250) # ylim 200, more than 95% data are included
)
plot
plot.save("../results/paiso_wt_" + title + ".pdf", height=2, width=2)
```
