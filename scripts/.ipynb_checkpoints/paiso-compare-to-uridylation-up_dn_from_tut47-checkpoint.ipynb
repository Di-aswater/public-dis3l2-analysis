```python
# morgan uridylation dataset:
tail_seq_morgan = pd.read_csv("../raw/TAIL-seq_Morganetal2017.csv")
tail_seq_morgan

# read in PAIso-seq:
df_paiso = pd.read_csv("../raw/PAIsoSeq_Wu_2022.csv")
```

```python
# only use GV in tail-seq:
df_tail_seq_GV = tail_seq_morgan[tail_seq_morgan['cell_tissue']=='GV']
df_tail_seq_GV
```

```python
tail_up = df_tail_seq_GV[df_tail_seq_GV['gv_upregulated']=='yes']
tail_no = df_tail_seq_GV[df_tail_seq_GV['gv_upregulated']=='no']
tail_no.groupby('gene').count()
```

```python
# from df_paiso, subset using tail_up and tail_no.
df_paiso_tail_up = df_paiso[df_paiso['gene_name'].isin(tail_up['gene'])]
df_paiso_tail_up #7065

df_paiso_tail_no = df_paiso[df_paiso['gene_name'].isin(tail_no['gene'])]
df_paiso_tail_no #231646
```

```python
# the percentage of polyA and U-polyA in genes of tail UACG
df_tail_seq_GV_U = df_tail_seq_GV[df_tail_seq_GV['modification_type_short']=='T']
df_tail_seq_GV_U #8119 transcripts

df_tail_seq_GV_A = df_tail_seq_GV[df_tail_seq_GV['modification_type_short']=='A']
df_tail_seq_GV_A #2855437 transcripts

df_tail_seq_GV_C = df_tail_seq_GV[df_tail_seq_GV['modification_type_short']=='C']
df_tail_seq_GV_C #6666 transcripts

df_tail_seq_GV_G = df_tail_seq_GV[df_tail_seq_GV['modification_type_short']=='G']
df_tail_seq_GV_G #17576 transcripts
```

```python
gene_oligoU = df_tail_seq_GV_U['gene'].drop_duplicates().tolist()
len(gene_oligoU) #2614

gene_oligoA = df_tail_seq_GV_A['gene'].drop_duplicates().tolist()
len(gene_oligoA) #12911

gene_oligoC = df_tail_seq_GV_C['gene'].drop_duplicates().tolist()
len(gene_oligoC) #2204

gene_oligoG = df_tail_seq_GV_G['gene'].drop_duplicates().tolist()
len(gene_oligoG) #3226
```

```python
# for all gene_oligoU, get their type:
df_gene_oligoU_type = df_paiso[df_paiso['gene_name'].isin(gene_oligoU)]
df_gene_oligoU_type #112869
df_gene_oligoU_type['mod']='oligoU'

df_gene_oligoA_type = df_paiso[df_paiso['gene_name'].isin(gene_oligoA)]
df_gene_oligoA_type #238363
df_gene_oligoA_type['mod']='oligoA'

df_gene_oligoC_type = df_paiso[df_paiso['gene_name'].isin(gene_oligoC)]
df_gene_oligoC_type #110193
df_gene_oligoC_type['mod']='oligoC'

df_gene_oligoG_type = df_paiso[df_paiso['gene_name'].isin(gene_oligoG)]
df_gene_oligoG_type #138537
df_gene_oligoG_type['mod']='oligoG'


df_oligoUACG = pd.concat([df_gene_oligoU_type,df_gene_oligoA_type,df_gene_oligoC_type,df_gene_oligoG_type])
df_oligoUACG
```

```python
info_df_oligoUACG = df_oligoUACG.groupby(['Genotype','mod','type'])['gene_ID'].nunique()
info_df_oligoUACG
```
