```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import plotnine as p9
%pylab inline
```

```python
# to study the the U-polyA vs polyA ratio in each oocyte.
```

```python
# read in df_paiso, containing all 272526 CCS reads from all paiso seq results.
df_paiso = pd.read_csv("../raw/PAIsoSeq_Wu_2022.csv",index_col=0)
df_paiso
```

```python
df_paiso['sampleID'] = df_paiso['Genotype']+'_' + df_paiso['oocyte_ID'].astype(str)
df_paiso
```

```python
# get the PC1 genes:
df_PC = pd.read_csv("../../results/PC1_from_rM_seq.csv")
df_PC
list_top100 = df_PC.loc[:500,'Unnamed: 0'].tolist()
# list_top100
# df_paiso[df_paiso['gene_name']=='Dis3l2']
```

```python
# visualize the top100 genes:
df_paiso_top100 = df_paiso[df_paiso['gene_ID'].isin(list_top100)]
df_paiso_top100
```

```python
df_paiso_top100.groupby(['Genotype','oocyte_ID','type']).count()
```

```python
df_paiso_top100['gene_ID'].nunique() #57
# plot each gene in ctr and cKO
info_df_paiso_top100 = df_paiso_top100.groupby(['gene_ID','Genotype','oocyte_ID']).count().reset_index()
info_df_paiso_top100

info_df_paiso_top100
# plot the count
title = 'info_df_paiso_top100'
plot = (
    p9.ggplot(info_df_paiso_top100, 
                  p9.aes(x='Genotype',y='CCS_ID',color='Genotype'))

    + p9.geom_jitter()
    + p9.geom_boxplot(outlier_shape='',alpha=0.3)
#     + p9.geom_point()
    + p9.theme(void)
    + p9.labs(title = title)
#     + p9.facet_wrap(['Genotype','oocyte_ID'])
    + p9.theme(axis_text_x=p9.element_text(angle=90))
#     + p9.xlim(0,250) # ylim 200, more than 95% data are included
)
plot
```

```python
# transcript number in each oocyte:
info_paiso = df_paiso.groupby(['Genotype','oocyte_ID','type']).count().reset_index()
info_paiso = info_paiso[['Genotype','oocyte_ID','type','CCS_ID']]
info_paiso.columns = ['Genotype','oocyte_ID','type','count']
info_paiso_U = info_paiso[info_paiso['type']=='U_polyA_RNA']
info_paiso_pa = info_paiso[info_paiso['type']=='polyA_RNA']

info_paiso_U
info_paiso_pa

info_paiso2 = pd.merge(left=info_paiso_U, right=info_paiso_pa, 
                       left_on=['Genotype','oocyte_ID'],
                      right_on=['Genotype','oocyte_ID'])
info_paiso2['Uvspa']= info_paiso2['count_x']/info_paiso2['count_y']
info_paiso2['id'] = info_paiso2['Genotype'] + '_' + info_paiso2['oocyte_ID'].astype('str')
info_paiso2['sum'] = info_paiso2['count_x'] + info_paiso2['count_y']

thre=1
info_paiso2 = info_paiso2[info_paiso2['sum']>thre]
info_paiso2
```

```python
# plot the ratio
title = 'U_polyA_vs_polyA_transcript'
plot = (
    p9.ggplot(info_paiso2, 
                  p9.aes(x='Genotype',y='Uvspa'))
    + p9.geom_boxplot(outlier_color='gray')
    + p9.geom_jitter(p9.aes(color='oocyte_ID'))
    + p9.theme(void)
    + p9.labs(title = title)
#     + p9.xlim(0,250) # ylim 200, more than 95% data are included
)
plot
# plot.save("../results/paiso_cKO_" + title + ".pdf", height=2, width=2)
```
