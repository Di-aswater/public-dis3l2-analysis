```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import plotnine as p9
%pylab inline
```

```python
# read in df_paiso, containing all 272526 CCS reads from all paiso seq results.
df_paiso = pd.read_csv("../raw/PAIsoSeq_Wu_2022.csv",index_col=0)
df_paiso
```

```python
# only select ctr group:
df_paiso_cKO = df_paiso[df_paiso['Genotype']=='cKO']
df_paiso_cKO

```

```python
# plot the tail vs pA:
title = 'tail_vs_pA_of_all_RNA'
plot = (
    p9.ggplot(df_paiso_cKO, 
                  p9.aes(x='tail_len',y='polyA_len'))
    + p9.geom_point(p9.aes(color='type'),alpha=0.1)
    + p9.theme(void)
    + p9.labs(title = title)
    + p9.coord_fixed()
    + p9.geom_abline(slope=1,color='gray',linetype="dashed",size=2)

)
plot
plot.save("../results/paiso_cKO_" + title + ".tiff", height=4, width=4,dpi=300)
```
