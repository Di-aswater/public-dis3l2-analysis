```python
import pandas as pd
import os
import glob
import matplotlib.pyplot as plt
import plotnine as p9
%pylab inline
```

```python
# purpose: to plot the variance of genes having U-polyA and not having U-polyA
# variance defination: the (value-mean)/mean.
```

```python
# read in df_paiso, containing all 272526 CCS reads from all paiso seq results.
df_paiso = pd.read_csv("../raw/PAIsoSeq_Wu_2022.csv",index_col=0)
df_paiso
# only select ctr group:
df_paiso_wt = df_paiso[df_paiso['Genotype']=='ctr']
df_paiso_wt

df_paiso_cKO = df_paiso[df_paiso['Genotype']=='cKO']
df_paiso_cKO

```

```python
df_use = df_paiso_wt.copy()
# genes that have U-polyA-RNA: U_genes; genes not having U-polyA RNA as not_U_genes
U_genes = df_use[df_use['type']=='U_polyA_RNA']['gene_ID'].unique()
U_genes
df_U_genes = df_use[df_use['gene_ID'].isin(U_genes)]
df_U_genes #70596 lines, 1862 genes
df_U_genes['gene_type'] = 'U_genes'

df_not_U_genes = df_use[-df_use['gene_ID'].isin(U_genes)]
df_not_U_genes #81127, 7489 genes
df_not_U_genes['gene_type'] = 'not_U_genes'

```

```python
# define the normalized variance of polyA len.
df_U_genes['mean_pA'] = df_U_genes['polyA_len'].mean()
df_U_genes['var'] = ((df_U_genes['polyA_len']- df_U_genes['mean_pA'])/df_U_genes['mean_pA']).abs()

df_U_genes

df_not_U_genes['mean_pA'] = df_not_U_genes['polyA_len'].mean()
df_not_U_genes['var'] = ((df_not_U_genes['polyA_len']- df_not_U_genes['mean_pA'])/df_not_U_genes['mean_pA']).abs()

df_not_U_genes
```

```python
df_merge = pd.concat([df_U_genes,df_not_U_genes])
df_merge
```

```python
plot = (
    p9.ggplot()
#     + p9.geom_boxplot(df_merge, 
#                   p9.aes(x='gene_type', y= 'var', color='gene_type'))
    + p9.geom_density(df_merge, p9.aes(x='var', color='gene_type'))
    + p9.theme(void)
    + p9.theme(axis_text_x = p9.element_text(angle=90)) 
    + p9.xlim(-1,3)
    + p9.geom_vline(xintercept=df_U_genes['var'].median(),color='red')
    + p9.geom_vline(xintercept=df_not_U_genes['var'].median(),color='green')
    
)
plot
plot.save("../results/paiso_polyA_dynamics_ctr_median.pdf",height=2, width=2)
```
